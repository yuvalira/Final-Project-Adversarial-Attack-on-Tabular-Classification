### 3. Correlation Analysis
To better understand how variables relate to income, we compute a correlation matrix.  
- Values close to +1 indicate a strong positive correlation.
- Values close to -1 indicate a strong negative correlation.
- Values near 0 suggest no strong relationship.



This heatmap visualizes the pairwise associations between features in the dataset using different statistical measures based on feature types.

* Pearson correlation is used for numerical–numerical pairs, measuring linear relationships.

* Cramér’s V is used for categorical–categorical pairs, capturing association strength regardless of category order.

* Mutual Information (MI) is used for numerical–categorical pairs, measuring both linear and non-linear dependencies.
To ensure comparability across metrics, MI values are normalized to the [0, 1] range.

The resulting matrix provides a unified view of feature interactions across the dataset.

```python
# --- Cramér’s V ---
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    rcorr = r - ((r - 1)**2) / (n - 1)
    kcorr = k - ((k - 1)**2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# --- Full mixed correlation matrix with normalization and plotting ---
def full_mixed_correlation(df, target_col):
    categorical_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
    numerical_cols = df.select_dtypes(include=["number"]).columns.tolist()
    all_cols = categorical_cols + numerical_cols
    corr_matrix = pd.DataFrame(np.zeros((len(all_cols), len(all_cols))), index=all_cols, columns=all_cols)

    # Label encode categorical columns for MI
    le_df = df.copy()
    label_encoders = {}
    for col in categorical_cols:
        le = LabelEncoder()
        le_df[col] = le.fit_transform(df[col])
        label_encoders[col] = le

    # --- Compute correlations ---
    for col1 in all_cols:
        for col2 in all_cols:
            if col1 == col2:
                corr_matrix.loc[col1, col2] = 1.0
            elif col1 in numerical_cols and col2 in numerical_cols:
                # Pearson correlation
                corr = df[col1].corr(df[col2])
                corr_matrix.loc[col1, col2] = corr
            elif col1 in categorical_cols and col2 in categorical_cols:
                # Cramér’s V
                confusion_mat = pd.crosstab(df[col1], df[col2])
                corr = cramers_v(confusion_mat)
                corr_matrix.loc[col1, col2] = corr
            elif ((col1 in numerical_cols and col2 in categorical_cols) or
                  (col2 in numerical_cols and col1 in categorical_cols)):
                # Mutual Information
                x = le_df[[col1]].values if col1 in numerical_cols else le_df[[col2]].values
                y = le_df[col2].values if col1 in numerical_cols else le_df[col1].values
                mi = mutual_info_classif(x, y, discrete_features=False)
                corr_matrix.loc[col1, col2] = mi[0]
                corr_matrix.loc[col2, col1] = mi[0]

    # --- Normalize MI values only (values > 1.0) ---
    mi_mask = corr_matrix > 1.0
    max_mi = corr_matrix[mi_mask].max().max()
    if max_mi > 0:
        corr_matrix[mi_mask] = corr_matrix[mi_mask] / max_mi

    # --- Top 5 features most associated with target_col ---
    scores = {}
    for col in all_cols:
        if col == target_col:
            continue
        if col in categorical_cols and target_col in categorical_cols:
            mat = pd.crosstab(df[col], df[target_col])
            scores[col] = cramers_v(mat)
        elif col in numerical_cols and target_col in numerical_cols:
            scores[col] = df[col].corr(df[target_col])
        else:
            x = le_df[[col]].values
            y = le_df[target_col].values
            mi = mutual_info_classif(x, y, discrete_features=False)
            scores[col] = mi[0]

    top_features = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:5]
    print("Top 5 features most associated with '{}':".format(target_col))
    for feature, score in top_features:
        print(f"{feature}: {score:.3f}")

    # --- Plot heatmap ---
    plt.figure(figsize=(10, 8))
    sn.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=0, vmax=1, annot_kws={"size": 12})
    plt.title("Feature Association Matrix (Pearson, Cramér’s V, MI normalized)", fontsize=16)
    plt.xticks(rotation=45, ha='right', fontsize=14)
    plt.yticks(fontsize=14)
    plt.tight_layout()
    plt.show()

    return corr_matrix


corr_matrix = full_mixed_correlation(df, target_col="income")
```

Since machine learning models work with numerical data, we need to convert categorical variables into numeric values. We use one-hot encoding (pd.get_dummies) for this purpose.

```python

\
df_numeric = df.copy()

# Convert the income column into a binary numeric variable ('<=50K' -> 0, '>50K' -> 1)
if df_numeric['income'].dtype == 'object':
    le = LabelEncoder()
    df_numeric['income'] = le.fit_transform(df_numeric['income'])

# Convert other categorical features to numeric
categorical_columns = df_numeric.select_dtypes(include=['object']).columns
df_numeric = pd.get_dummies(df_numeric, columns=categorical_columns, drop_first=True)

df_numeric
```

```python
# Calculate the correlation
correlation_matrix = df_numeric.corr()['income'].sort_values(ascending=False)

# Display the top 10 correlation values
print("Top 10 features most correlated with 'income':")
print(correlation_matrix[1:11])

# Extract the feature names and their correlations
features = correlation_matrix.index
correlations = correlation_matrix.values

# Plot
plt.figure(figsize=(15, 8))
sn.barplot(x=features, y=correlations, color="lightsteelblue")

# Adjust x-axis for readability
plt.xticks(rotation=90, fontsize=8)
plt.title('Feature Correlation with Income', fontsize=14)
plt.ylabel('Correlation', fontsize=12)
plt.xlabel('Features', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

Now, we use a heatmap to visualize the relationships of the top 10 features most correlated with 'income'

```python
# Compute the full correlation matrix
correlation_matrix2 = df_numeric.corr()

# Ensure 'income' exists in correlation_matrix
if "income" in correlation_matrix2.columns:
    # Select the top 10 most correlated features with income
    top_features = correlation_matrix2["income"].abs().sort_values(ascending=False).index[:10]

    # Create a heatmap with only those features
    plt.figure(figsize=(10, 6))  # Adjust figure size
    sn.heatmap(df_numeric[top_features].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

    plt.title("Top 10 Feature Correlations with Income")
    plt.show()
else:
    print("Error: 'income' column is not in correlation_matrix. Check if df_numeric.corr() was computed correctly.")
```
